## Definição e seletores

Após a criação do HTML a necessidade de formatar as páginas ficou evidente, assim, em 1996, foi criada a linguagem de estilo que conhecemos por CSS.<p>
A sintaxe é bem simples e pode ser aplicada com a frase "você cria reggras de estilo para elementos ou grupos de elementos.<p>
Vamos usar um elemento HTML que vimos anteriormente, a âncora _a_, para exemplificar.<P>
Uma reggra CSS é representada por um seletor ou um grupo de seletores, no nosso caso é o _a_, então dentro de um par de chaves adicionamos as declarações, no exemplo acima estamos alterando cor e tamanho dessa âncora, as declarações são formadas por uma propriedade e um valor.<p>
Percebam que podemos colocar vários seletores em uma regra separando-os por vírgula.<p>
E há um último detalhe nesse exemplo: a pseudo-classe. Elementos HTML sofrem alterações causadas pela interação do usuário, como mover o mouse por cima ou clicar nesse elemento.<p>
O _a:hover_ do exemplo significa que a Âncora também terá essa aparência quando o usuário passar o mouse por cima de um _hiperlink_.


## ID X Classe

No exemplo anterior criamos uma regra que altera um elemento HTML diretamente, mas isso significa que todos os elementos _a_ ficarão com aquela aparência, e normalmente temos sites mais complexos que precisam de várias regras diferentes para elementos iguais.<p>
Para ficar mais tangível vamos relembrar um pouco o site que começamos a fazer no módulo passado, ele tinha vários elementos header , mas não vamos querer que o header principal tenha a mesma formatação que o header de uma postagem, é aí que entram as IDs e Classes.<p>
O seletor que vimos no primeiro exemplo é um seletor de tipo, pois ele reprsenta um elemento HTML, e com IDs e Classes podemos representar qualquer tipo de elmento mas há algumas diferenças entre eles:

ID: é representado pelo símbolo # (hash) seguido de um nome para esse ID.<p>
Classe: a classe é representada de forma parecida do ID, mas é precedida de um ponto ao invés de hash.<p>
E a diferença mais importante entre eles é a fforma como deem ser usados: o ID pode ser usado uma vez em uma página HTML enquanto a Classe não tem restrições.

## Box-model

Quando estamos criando o layout de um site o navegador representa cada elemento HTML como uma caixa retangular, isso pe o box-model. E com o CSS nós alteramos a aparência dessa caixa (largura, altura, cor de fundo, etc). Essa caixa é composta por 4 áreas:  <p>
- conteúdo
- padding
- borda
- margem

_content_ - conteúdo é o que o seu bloco representa, um texto, uma imagem, um vídeo.<p>
_padding_ - é o espaçamento entre as bordas e o conteúdo, a diferença para as margens é que declarações da imagem de fundo funcionem nele<p>
_border_ - bordas<p>
_margin_ - margens são espaçamentos entre elementos<p>





## Exercício

Para enxergarmos o _box-model_ vamos adicionar cores e bordas a alguns elementos.<p>
Primeiro adicionaremos uma cor de fundo para a visualização ficar mais fácil, usaremos a propriedade _background_ com o valor _#fcfcfc_ no elemento _body_.

```
body{
    background: #fcfcfc
}

```

Depois vamos adicionar uma classe ao _article_. pode ser _.post_, e então vamos colocar a cor branca de fundo com a propriedade _background_ e o valor _#fff_.<p>
Agora conseguimos enxergar o _content_ do _box-model_.<p>

```
...
        <article class="post">
            <header><h3 class = "post_title">Posts #1</h3>
                    <img src="foto_istock.jpg" alt="Ilustração programação HTML" class = "post_image"> 
            </header>
            ...
```
Vamos adicionar um _padding_ de 10 pixels neste mesmo _article_. Perceberam o espaço que surgiu em volta do nosso conteúdo?<p>

```
.post{
    background: #ffff;
    padding: 15px;
}
```

Agora adicionamos uma borda mais escura a ele com a propriedade _border_. Vou falar mais detalhadamente sobre border mais à frente, mas por enquanto vamos deixar essa borda com 3 pixels de largura, o contorno sólido e a cor azul.<p>

```
.post{
    background: #ffff;
    padding: 15px;
    border: 3px solid blue;
}
```
E por último vamos adicionar uma margem do lado de fora do post com a propriedade _margin_ e o valor 10 pixels.

```
.post{
    background: #ffff;
    padding: 15px;
    border: 3px solid blue;
    margin-bottom: 10px;
}
```
E agora inspecionando o nosso elemento, conseguimos todas aquelas camadas citadas antes, o conteúdo em azul, o _padding_ em verde, as bordas em marrom e as margens em laranja.

# Estilizando elementos

## Padding e Margin
Anteriormente usamos o _padding_ e o _margin_ da forma mais básica, mas eles são mais poderosos que isso. Se quisermos atribuir tamanhos diferentes para cada lado do _box_, nós podemos e temos 3 formas de fazer isso.<p>

A primeia é colocando um valor para as partes superior e inferior e depois para o lado esquerdo e direito.<p>

```
.post{
    padding: 10px 5 px; 
     } 

```

O valor de 10 pixels refere-se ao eixo _y_, ou as partes superior e inferior, e os 5 pixels referem-se aos lado esquerdo e direito.<p>

A segunda forma é dando valores para cada lado do _box_.<p>

```
.post{
    padding: 15px 10px 5px 0;
}
```
Então começamos pelo _topo_ com 15 pixels, passamos para o lado _direito_ com 10 pixels, depois para a parte _inferior_ com 5 pixels e por último o lado _esquerdo_ com 0, e sempre nessa ordem:<p>
- topo
- direito
- inferior
- esquerdo<p>

Uma boa dica também é que quando o valor for 0 não precisamos colocar a unidade.<p>

A terceira forma é com as propriedades específicas para cada lado, até agora tínhamos visto atalhos para essas propriedades.<p>
```
.post{
    padding-top: 15px;
    padding-right: 10px;
    padding-bottom: 5px;
    padding-left: 0;
}
```

Essa opção é mais usada quando temos o mesmo valor para os 3 lados, e o quarto precisa ter um valor diferente, então usamoso _padding_ com apaenas um valor e uma das opções para representar o lado diferente.<p>

## Border
Vimos que a propriedade _border_ pode ter 3 valores:
- lagura
- cor
- estilo<p>

Mas existem algumas particularidades nisso.<p>
A largura pode ser usada com várias unidades como _px_, e em _mm_. A cor pode er atribuída pelo nome ou peos um código hexadecimal, assim como fizemos com o _background_ e o estilo é representado por palavra-chave, vamos ver algumas delas:
- solid: mostra uma borda simples e reta;
- dotted: são bolinhas com uma pequeno espaçamento entre elas;
- dashed: forma uma linha tracejada.<p>

E aproveitando que mostrei esse código temos que falar sobre como separar a estilização dis lados de uma borda.<p>
E se você não quiser usar a propriedade _border_ existem as propriedades específicas para cada espectro de um borda, são elas _border-width_ para a largura, _border_color_ para a cor e _border-style_ par o estilo.<p>
Aqui temos o  mesmo código anterior de duas formas diferentes, a primeira com o atalho  _border_ e a segunda com cada propriedade específica.
E depois disso podemos juntar os lados com os aspectos de uma borda e criari uma regra mais específica ainda.<p>

## Border-radius
E a última propriedade é o _border-radius_, ele permite arredondar os cantos de um elemento. Podemos usar várias unidades, mas as mais comuns são os pixels e a porcentagem.<p>

Colocando apenas um valor mudamos todos os cantos do elemento, mas seguindo aquela mesma ordem que vimos no _padding_ e _margin_: topo, direita, inferior e esquerda - conseguimos alterar cada canto separadamente.<p>

## Estilizando textos
Já sabemos que podemos mudar de cor e tamanho algumas fontes e agora vamos nos aprofundar nisso.<p>

### font-family
Com o _font-family_ podemos alterar a fonte dos nossos textos, como uma fonte da intyernet ou uma que esteja instalada no seu computador, mas vamos nos ater às fontes seguras, chamadas de web safe fonts.<p>
Essas fontes são chamadas assim pois são encontradsa em quase todos os sistemas e pode ser usadas sem preocupação.<p>

```
#title{
    font-family: Verdana;
}

.post_title{
    font-family: Verdana, Arial;
}

```
(no caso das duas fontes informadas, a segunda entra quando a primeira não da certo, a segunda é um backup)

### font-size
O _font-size_ nos ajuda a mudar o tamanho do texto, existem algumas unidades de medida para ele, mas por enquanto os pixels são suficientes para nós.<p>

```
#title{
    font-size: 30px
}

.post_title{
    font-size: 18px
}

```
### font-style
Usamos o _font-style_ para tornar um texto itálico, na maioria das vezes você pode usar apenas o valor _italic_ para ele, mas se precisar tirar o itálico de um texto voc~e pode usar o valor _normal_.<p>

```
#title{
    font-style: normal;
}

.subtitle{
    font-style: italic;
}
```
### font-weight

```
#title{
    font-weight: normal;
}
.subtitle{
    font-weight: bold;
}
```

### text-transform

```
#title{
    text-transform: uppercase:
}

.subtitle{
    text-transform: lowercase;
}

.post_title{
    text-transform: capitalize;
}
```

### text-decoration

```
#title{
    text-decoration: underline;
}

.subtitle{
    text-decoration: overline;
}

.post_title{
    text-decoration: line-through
}
```

### list-style-type

```
ul {
    list-style type: square;
}

ol {
    list-style-type: upper-roman;
}

ul {
    list-style-type: "\1f44d";
}
```

### list-style-image

```
ul {
    list-style-image: url("rocket.png");
}

```
